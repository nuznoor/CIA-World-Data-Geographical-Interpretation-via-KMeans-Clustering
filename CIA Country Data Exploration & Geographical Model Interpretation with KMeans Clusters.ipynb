## Data Preparation

```python
df.isna().sum()
```

```python
### Countries that have NaN for agriculture are mostly islands, so let's put 0s for those
df[df['Agriculture'].isnull()]['Country']
```

```python
df['Agriculture'] = df['Agriculture'].fillna(0)
```

```python
df.isna().sum()
```

```python
## Filling in missing climate values with mean climate values for region
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
df.isna().sum()
```

```python
## Filling in missing literacy values with mean literacy values for region
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
df.isna().sum()
```

```python
## Filling in missing literacy values with mean literacy values for region
df['GDP ($ per capita)'] = df['GDP ($ per capita)'].fillna(df.groupby('Region')['GDP ($ per capita)'].transform('mean'))
df.isna().sum()
```

```python
## Dropping rest of the rows with missing values
df = df.dropna()
df.isna().sum()
```

## KMeans

```python
from sklearn.cluster import KMeans
```

```python
## Country Column is unique identifier, we don't need in clustering
X = df.drop(['Country'], axis=1)
## Creating Dummies for Categorical Variables
X = pd.get_dummies(df)
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
scaled_X
```

### Finding Optimal K-Value

```python
ssd = []

for k in range(2,50):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    
    ##.inertia_(does the sum of squared distances to its cluster center)
    ssd.append(model.inertia_)
```

```python
plt.figure(figsize=(10,6), dpi=200)
plt.plot(range(2,50), ssd, 'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
## we will choose k=15 bc the drops after that starts leveling off
plt.figure(figsize=(10,6), dpi=200)

pd.Series(ssd).diff().plot(kind='bar')
```

### Final Model

```python
final_model = KMeans(n_clusters=15)
final_model.fit(scaled_X)
```

## Geographical Model Interpretation of Clusters
- viewing clusters on choropleth map

```python
import plotly.express as px
```

```python
country_code = pd.read_csv('country_iso_codes.csv')
country_code.head()
```

```python
iso_mapping = country_code.set_index('Country')['ISO Code'].to_dict()
```

```python
df['ISO Code'] = df['Country'].map(iso_mapping)
```

```python
df['Cluster']=final_model.labels_
```

```python
import plotly.express as px

fig = px.choropleth(df, geojson=df['Country'], locations=df['ISO Code'], color='Cluster',
                           color_continuous_scale="Turbo",
                           range_color=(0, 14),
                            hover_name='Country',
                           scope="world",
                           labels={'Cluster'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
